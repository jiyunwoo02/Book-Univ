# 1 데이터 불러오기

```python
import numpy as np
import pandas as pd
import re
from bs4 import BeautifulSoup
from konlpy.tag import Okt
```

```python
# 데이터 불러오기
# 도서 데이터
#df= pd.read_excel('가톨릭대 인기도서 50권.xlsx')
df= pd.read_csv('한국외대_크롤링_완료.csv')
df
```

```python
# 필요 컬럼만 남기기
# '구름'과 '양' 컬럼 삭제
df = df.drop(['목차', '출판사 서평', '추천사'], axis=1)
```

# 2 특수문자 및 태그 / 영어 및 숫자 제거

```python
# 컬럼 확인
columns = df.columns

# 컬럼 출력
print(columns)
```

```python
# 특수 문자, 영어, 숫자, 개행 문자 제거 함수 정의
def remove_special_chars(text):
    if isinstance(text, str):
        soup = BeautifulSoup(text, "html.parser")
        cleaned_text = soup.get_text()
        cleaned_text = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ\s]', '', cleaned_text)
        cleaned_text = re.sub('[\r\n]+', ' ', cleaned_text)
        return cleaned_text
    else:
        return text

# 대출 횟수를 제외한 컬럼에 대해서 수행
# 특수 문자, 영어, 숫자, 개행 문자 제거할 컬럼들 리스트
columns_to_clean = ['도서명', '저자', '출판사', '발행연도', '대출횟수', '카테고리', '저자소개', '책소개']

# 각 컬럼에 특수 문자, 영어, 숫자, 개행 문자 제거 적용
for column in columns_to_clean:
    df[column] = df[column].apply(remove_special_chars)
```

# 4 맞춤법 및 띄어쓰기

```python
from hanspell import spell_checker

def preprocess_text(text):
    # 최대 499자씩 나누어 맞춤법 검사 수행
    chunks = [text[i:i+499] for i in range(0, len(text), 499)]
    corrected_chunks = []
    for chunk in chunks:
        spelled_text = spell_checker.check(chunk)
        corrected_chunks.append(spelled_text.checked)
    # 결과를 합치기
    checked_text = ''.join(corrected_chunks)
    return checked_text

# 전처리할 열 목록
columns_to_preprocess = ['저자소개', '책소개']

# 각 열에 대한 전처리 적용
for column in columns_to_preprocess:
    df[column] = df[column].apply(lambda x: preprocess_text(x) if isinstance(x, str) else x)
```

# 5 토큰화 및 형태소 분석

```python
new_df = df
```

```python
# Okt
tokenizer = Okt()

# Tokenization and morphological analysis 함수
def tokenize_and_analyze(text):
    if isinstance(text, str):
        tokens = tokenizer.morphs(text)  # Morphological analysis
        return tokens
    else:
        return []

# 토큰화 및 형태소 분석을 적용할 컬럼들 리스트
columns_to_tokenize = ['저자소개', '책소개']

# 각 컬럼에 토큰화 및 형태소 분석 적용
for column in columns_to_tokenize:
    new_df[column] = new_df[column].apply(tokenize_and_analyze)
```

# 6 단어 정규화

```python
from konlpy.tag import Komoran
```

```python
def normalize_text(text):
    komoran = Komoran()
    tokens = komoran.morphs(text)
    normalized_tokens = [komoran.normalize(token) for token in tokens]
    normalized_text = ' '.join(normalized_tokens)
    return normalized_text
```

```python
# 정규화할 열 목록
columns_to_normalize = ['저자소개', '책소개']
```

```python
# 각 열에 대한 정규화 적용
for column in columns_to_normalize:
    new_df[column] = new_df[column].apply(lambda x: normalize_text(x) if isinstance(x, str) else x)
```

# 7 불용어 처리

```python
import nltk
from nltk.corpus import stopwords
```

```python
# 불용어 제거 함수
def remove_stopwords(tokens):
  # 불용어 리스트
  with open('stopwords.txt', 'r', encoding='utf-8') as file:
        korean_stopwords = [line.strip() for line in file]
  filtered_tokens = [token for token in tokens if token not in korean_stopwords]
  return filtered_tokens

# 불용어 제거를 적용할 컬럼
columns_to_clean =['저자소개', '책소개']

# 컬럼 별로 불용어 제거
for column in columns_to_tokenize:
    new_df[column] = new_df[column].apply(remove_stopwords)
```
