# 정규표현식

## 1

```python
import requests
import re

# 웹 페이지의 HTML 코드
response = requests.get('http://www.yes24.com/Product/Goods/118542760')
html = response.text

# 정규 표현식을 사용하여 웹 페이지에서 링크를 추출
links = re.findall('<a href="(.*?)">', html)

# 추출된 링크를 출력
for link in links:
    print(link)
```

## 2 제목

```python
import requests
import re

url = "http://www.yes24.com/Product/Goods/118542760"
response = requests.get(url)
html = response.text

# 책 제목 추출
title_pattern = r'<meta name="title" content="(.*?)"/>'
title = re.findall(title_pattern, html)[0]

# 추출된 책 제목을 출력
print(title)
```

## 3 작가명

```python
# 작가 이름 추출
author_pattern = r'<meta name="author" content="(.*?)" />'
result = re.findall(author_pattern, html)[0]

print(result)
```

# 4 책소개

```python
# 책 소개 추출
intro_pattern = r'<meta name="description" content="(.*?)"/>'
result = re.findall(intro_pattern, html)[0]

print(result)
```

# 5 출판사리뷰

```python
# 출판사리뷰 추출
review_pattern = r'<div id="infoset_pubReivew" class="gd_infoSet gd_infoSetCrop">\s+<div class="tm_infoSet">\s+<h4 class="tit_txt">출판사 리뷰</h4>\s+</div>\s+<div class="infoSetCont_wrap">\s+<div class="infoWrap_txt">\s+<b>(.*?)</b>'
result = re.findall(review_pattern, html, re.DOTALL)[0]

result_text = re.sub(r'<br\s*/?>', ' ', result)

print(result_text.strip())
```

# 1. 웹 문서 전체 가져오기
- 인터파크도서 홈페이지 > <우리에게는 온기가 있기에> 책 클릭

```python
!pip install requests beautifulsoup4
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

html = urlopen('https://book.interpark.com/product/BookDisplay.do?_method=detail&sc.shopNo=0000400000&sc.saNo=001&sc.prdNo=356729411&bid1=17_main&bid2=Field&bid3=tab01&bid4=prd_02')
soup = BeautifulSoup(html, 'html.parser')

print(soup)
```

# 2. 웹 타이틀 가져오기

```python
print(soup.head.title)
```

# 3. 모든 메타 데이터 내용 가져오기

```python
for meta in soup.head.find_all('meta'):
    print(meta.get('content'))
```
