```python
import numpy as np
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup


# URL 정보를 데이터프레임에 저장하는 함수
def save_url_info_to_dataframe(url, df):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'lxml')

    # 책 제목
    tag_name = '#yDetailTopWrap > div.topColRgt > div.gd_infoTop > div > h2'
    books = soup.select(tag_name)
    book_title = books

    # 출판사
    publisher_tags = soup.select('span.gd_pubArea>span.gd_pub>a')
    if publisher_tags:
        publisher = publisher_tags
    else:
        publisher = 'NaN'
        
    # 발행일자
    #//*[@id="yDetailTopWrap"]/div[2]/div[1]/span[2]/span[3]
    date_tags = soup.select('div.topColRgt>div.gd_infoTop>span.gd_pubArea>span.gd_date')
    if date_tags:
        date = date_tags
    else:
        date = 'NaN'
    
    # 카테고리
    category_tags = soup.select('div.infoSetCont_wrap>dl.yesAlertDl>dd>ul.yesAlertLi>li>a')
    if category_tags:
        category = category_tags
    else:
        category = 'NaN'
    
    # 작가
    author_tags = soup.select('span.gd_auth > a')
    author = author_tags
    
    #작가 소개
        
    author_info_tags = soup.select('span.author_info.info_crop')
    if author_info_tags:
        author_info = author_info_tags
    else:
        author_info = 'NaN'


    # 책 소개
    introduce_tags = soup.select('div.infoWrap_txtInner')
    # book_introduce = introduce_tags
    
    if introduce_tags:
        book_introduce = introduce_tags
    else:
        book_introduce = 'NaN'
        

    # 목차
    info_tags = soup.select('#infoset_toc > div.infoSetCont_wrap > div.infoWrap_txt')
    # table_of_contents = info_tags
    
    if info_tags:
        table_of_contents = info_tags
    else:
        table_of_contents = 'NaN'
        

    # 출판사 리뷰
    review_pub_tags = soup.select('#infoset_pubReivew > div.infoSetCont_wrap > div.infoWrap_txt')
    # publisher_review = review_pub_tags
    if review_pub_tags:
        publisher_review = review_pub_tags
    else:
        publisher_review = 'NaN'

    # 추천평
    res_tags = soup.select('#infoset_nomiCmt > div.infoSetCont_wrap > div.infoWrap_txt')
    # recommendations = res_tags
    if res_tags:
        recommendations = res_tags
    else:
        recommendations = 'NaN'
        

    # 데이터프레임에 저장
    new_row = {
        #'URL': url,
        '도서명': book_title,
        '저자' : publisher,
        '발행연도' : date,
        '카테고리' : category,
        '저자': author,
        '저자소개' : author_info,
        '책소개': book_introduce,
        '목차': table_of_contents,
        '출판사서평': publisher_review,
        '추천사': recommendations
    }

    df1 = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    return df1
```

# 과기대 12개 파일 합치기 + 처리

```python
import pandas as pd
data6=pd.read_excel('과기대.xlsx')
data6=data6.sort_values('대출권수',ascending=False)
data6
```

```python
# 중복된 '도서명'을 가진 행들의 '대출권수' 합계 계산
data6_grouped = data6.groupby('도서명')['대출권수'].sum().reset_index()

# 대출권수를 정수로 변환
data6_grouped['대출권수'] = data6_grouped['대출권수'].astype(int)

# 대출권수를 기준으로 내림차순 정렬
data6_grouped = data6_grouped.sort_values('대출권수', ascending=False)
data6_grouped
```

```python
# 중복된 '도서명'을 가진 행들의 '대출권수' 합계 계산
data6_grouped = data6.groupby('도서명')['대출권수'].sum().reset_index()

# 저자, 출판사, 출판연도 열 추가
data6_grouped = pd.merge(data6_grouped, data6[['도서명', '저자', '출판사', '출판연도']].drop_duplicates(), on='도서명')

# 대출권수를 정수로 변환
data6_grouped['대출권수'] = data6_grouped['대출권수'].astype(int)

# 대출권수를 기준으로 내림차순 정렬
data6_grouped = data6_grouped.sort_values('대출권수', ascending=False)

data6_grouped['출판연도'] = data6_grouped['출판연도'].astype(int)

# 결과 출력
data6_grouped
```
