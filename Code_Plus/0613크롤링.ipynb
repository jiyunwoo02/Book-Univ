```python

# 기존의 데이터 프레임(data_U)에서 필요한 열만 추출하여 새로운 데이터 프레임 생성
df_U = data_U[['서명', '저자', '대출횟수']].copy()
import pandas as pd

# '도서명' 열의 이름을 '도서명_'으로 변경
df_U.rename(columns={'도서명': '도서명_'}, inplace=True)
df_U.rename(columns={'저자': '저자_'}, inplace=True)
# 새로운 데이터 프레임 출력
df_U.head(5)
```

```python
# URL 정보를 데이터프레임에 저장하는 함수
def save_url_info_to_dataframe(url, df):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'lxml')

    # 책 제목
    tag_name = '#yDetailTopWrap > div.topColRgt > div.gd_infoTop > div > h2'
    books = soup.select(tag_name)
    book_title = books

    # 출판사
    publisher_tags = soup.select('span.gd_pubArea>span.gd_pub>a')
    if publisher_tags:
        publisher = publisher_tags
    else:
        publisher = 'NaN'
        
    # 발행일자
    #//*[@id="yDetailTopWrap"]/div[2]/div[1]/span[2]/span[3]
    date_tags = soup.select('div.topColRgt>div.gd_infoTop>span.gd_pubArea>span.gd_date')
    if date_tags:
        date = date_tags
    else:
        date = 'NaN'
    
    # 카테고리
    category_tags = soup.select('div.infoSetCont_wrap>dl.yesAlertDl>dd>ul.yesAlertLi>li>a')
    if category_tags:
        #category = category_tags
        category = category_tags[1].text.strip()
    else:
        category = 'NaN'
    
    # 작가
    author_tags = soup.select('span.gd_auth > a')
    if author_tags:
        author = author_tags
    else:
        author = 'NaN'
    
    #작가 소개
    #/html/body/div[2]/div[7]/div/div[2]/div[9]/div[2]/div/div[2]/div[1]/span[4]
    
    author_info_tags = soup.select('span.author_info.info_origin')
    if author_info_tags:
        author_info = author_info_tags[0].text.strip()
    else:
        author_info = 'NaN'
        

    # 책 소개
    #td>div.infoWrap_txt>
    introduce_tags = soup.select('div.infoWrap_txtInner')
    # book_introduce = introduce_tags
    
    if introduce_tags:
        #book_introduce = introduce_tags
        book_introduce = introduce_tags[0].text.strip()
    else:
        book_introduce = 'NaN'
        

    # 목차
    info_tags = soup.select('#infoset_toc > div.infoSetCont_wrap > div.infoWrap_txt')
    # table_of_contents = info_tags
    
    if info_tags:
        table_of_contents = info_tags
    else:
        table_of_contents = 'NaN'
        

    # 출판사 리뷰
    review_pub_tags = soup.select('#infoset_pubReivew > div.infoSetCont_wrap > div.infoWrap_txt')
    # publisher_review = review_pub_tags
    if review_pub_tags:
        publisher_review = review_pub_tags
    else:
        publisher_review = 'NaN'

    # 추천평
    res_tags = soup.select('#infoset_nomiCmt > div.infoSetCont_wrap > div.infoWrap_txt')
    # recommendations = res_tags
    if res_tags:
        recommendations = res_tags
    else:
        recommendations = 'NaN'
        

    # 데이터프레임에 저장
    new_row = {
        #'URL': url,
        '도서명': book_title,
        '출판사' : publisher,
        '발행연도' : date,
        '카테고리' : category,
        '저자': author,
        '저자소개' : author_info,
        '책소개': book_introduce,
        '목차': table_of_contents,
        '출판사 서평': publisher_review,
        '추천사': recommendations
    }

    df1 = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    return df1
```
